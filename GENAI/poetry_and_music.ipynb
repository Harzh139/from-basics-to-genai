```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
model_name = 'gpt2'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
def gen_poem(prompt,max_length):
  #encoding
  input_ids = tokenizer.encode(prompt,return_tensors = 'pt')
  output = model.generate(
      input_ids,
      max_length = max_length,
      num_return_sequences = 1,
      no_repeat_ngram_size = 2,
      top_k = 50,
      top_p = 0.95,
      temperature = 0.7
  )
  generated_text = tokenizer.decode(output[0],skip_special_tokens=True)

  return generated_text
```

```python
prompt = 'Twinkle Twinkle little star'
max_length = 100
poem = gen_poem(prompt,max_length)
print(poem)
```

```python
from music21 import *
import random
```

```python
def gen_melody(length=8,scale = scale.MajorScale('C')):
  melody = stream.Stream()

  for _ in range(length):
    pitch = random.choice(scale.getPitches('C4','C5'))

    duration = random.choice([0.5,1,1.5,2])
    new_note = note.Note(pitch)
    new_note.duration.quarterLength = duration
    melody.append(new_note)
  return melody
```

```python
melody = gen_melody()
melody.show('midi')
```

```python
def poem_to_melody(poem, scale = scale.MajorScale("A")):

  words = poem.split()

  melody = stream.Stream()

  for word in words[:8]: # limit to 8 notes for simplicity

    pitch = scale.pitchFromDegree((len(word)%7) + 1)

    duration = len(word) * 0.25 # Quarter node for each letter

    new_note = note.Note(pitch)

    new_note.duration.quarterLength = duration

    melody.append(new_note)

  return melody
```

```python
poem = gen_poem("The wind whispers, ",max_length)

print("+++++++++++++++++POEM+++++++++++++++++++")
print(poem)

print("****************************************")

melody = poem_to_melody(poem)

melody.show('midi')
```
